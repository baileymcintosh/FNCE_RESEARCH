```python
# Cache round-trip test
test = pd.DataFrame({'a': [1, 2, 3]})
cache.write(test, 'test', 'sanity', 0, 0)
back = cache.read('test', 'sanity', 0, 0)
assert back.equals(test)
cache.invalidate('test', 'sanity', 0, 0)
print('Cache: OK')
print('All checks passed.')
```
