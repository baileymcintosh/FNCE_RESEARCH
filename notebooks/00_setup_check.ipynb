# Setup Check

Run this notebook after initial setup to verify:
1. The package is installed correctly
2. Environment variables are loaded
3. WRDS connection works
4. Cache paths are writable

```python
# 1. Package imports
import fnce_research
from fnce_research import config
print('Package loaded OK')
print(f'  REPO_ROOT   : {config.REPO_ROOT}')
print(f'  DATA_RAW    : {config.DATA_RAW}')
print(f'  RESULTS_DIR : {config.RESULTS_DIR}')
print(f'  WRDS_USER   : {config.WRDS_USERNAME}')
```

```python
# 2. WRDS connection
from fnce_research.wrds_conn import get_db

db = get_db()   # will prompt for password on first run, then save to ~/.pgpass
print('Connected to WRDS:', db)
```

```python
# 3. Quick query sanity check
df = db.raw_sql("""
    SELECT permno, date, ret
    FROM crsp.msf
    WHERE date = '2023-12-29'
    LIMIT 10
""", date_cols=['date'])
print(f'{len(df)} rows returned')
df.head()
```

```python
# 4. Cache write test
from fnce_research.data import cache
import pandas as pd

test_df = pd.DataFrame({'a': [1, 2, 3]})
cache.write(test_df, 'test', 'sanity', 0, 0)
back = cache.read('test', 'sanity', 0, 0)
assert back.equals(test_df), 'Cache round-trip failed!'
cache.invalidate('test', 'sanity', 0, 0)
print('Cache read/write/delete: OK')
```

```python
# 5. Close connection
from fnce_research.wrds_conn import close_db
close_db()
print('All checks passed.')
```
